```python
import mysql.connector
import matplotlib.pyplot as plt

db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Meesam_rizvi69",
    database="weather_coverage"
)

cursor = db.cursor()
cursor.execute("SELECT temp, humidity, wind_speed ,time FROM data")
rows = cursor.fetchall()
cursor.close()
db.close()

temp = []
humidity = []  
wind_speed = []
time = []


for row in rows:
    temp.append(row[0]) 
    humidity.append(row[1])  
    wind_speed.append(row[2]) 
    time.append(row[3])

humidity.sort()
temp.sort()
wind_speed.sort()
time.sort()
```

```python
plt.plot(time,temp)
plt.title("Temparture")
plt.xlabel("time")
plt.ylabel("temp")
plt.grid(True)
plt.show()
```

```python
plt.plot(time,humidity)
plt.title("Humidity")
plt.xlabel("time")
plt.ylabel("humidity")
plt.grid(True)
plt.show()
```

```python
plt.plot(time,wind_speed)
plt.title("Wind Speed")
plt.xlabel("time")
plt.ylabel("speed")
plt.grid(True)
plt.show()
```
